```python
response = requests.get('https://www.efcc.gov.ng/efcc/')
html_content = response.text
soup = BeautifulSoup(html_content,'lxml')
print(soup.prettify())
```

```python
soup.find('strong').text
```

```python
paragraph = soup.find_all('p')

for para in paragraph:
    print(para.text)    
```
