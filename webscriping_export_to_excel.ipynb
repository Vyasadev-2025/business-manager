```python
url = 'https://jiji.ng/vehicles'
response = requests.get(url)
html_content = response.text
jiji_soup = BeautifulSoup(html_content, 'lxml')

data = []

item_collection = jiji_soup.find_all('div', class_ = 'masonry-column')
for items in item_collection:
    jiji_items = items.find_all('div', class_ = 'masonry-item')
    for goods in jiji_items:
        product_name = goods.find('div', class_ = 'qa-advert-title')
        Name = product_name.text.strip() if product_name else "NA"
   
        product_price = goods.find('div', class_ = 'qa-advert-price')
        Price = product_price.text.strip() if product_price else "NA"

        product_describe = goods.find('div', 'b-list-advert-base__description-text')
        Description = product_describe.text.strip() if product_describe else "NA"

        product_img = goods.find('img')
        Image_link = product_img.get('src') if product_img else "NA"

        goods_info = {
             'Item': Name,
             'Price': Price,
             'Description': Description,
             'Image_link': Image_link
        }
        


        data.append(goods_info)
```

```python
df = pd.DataFrame(data)
df.head(3)
```

```python
df['Brand'] = df['Item'].apply(lambda x: str(x).split(' ')[0])
```

```python
df.to_excel('jiji_data.xlsx', index=False)
```
