```python
edge_service = Service(executable_path="C:/Users/HP/Downloads/edgedriver_win64/msedgedriver.exe")  # Update with your path        
driver = webdriver.Edge(service=edge_service) 
url = 'https://www.jiji.ng/vehicles'  # URL of the website you want to open     
driver.get(url)  # Open a website  

jiji_data = []

item_col = driver.find_elements(By.CLASS_NAME, 'masonry-column')

for item in item_col:
    goods = item.find_elements(By.CLASS_NAME, 'masonry-item')
    for good in goods:
        product_name = good.find_element(By.CLASS_NAME, 'qa-advert-title')
        name = product_name.text.strip() if product_name else 'NA'

        product_price = good.find_element(By.CLASS_NAME, 'qa-advert-price')
        price = product_price.text.strip() if product_price else 'NA'
        
        product_description = good.find_element(By.CLASS_NAME, 'b-list-advert-base__description-text')
        description = product_description.text.strip() if product_description else 'NA'
        
        #product_location = good.find_element(By.CLASS_NAME, 'b-list-advert_region_text')
        #location = product_location.text.strip() if product_location else 'NA'
        
        product_img = good.find_element(By.TAG_NAME,'img')
        image = product_img.get_attribute('src') if product_img else 'NA'

        try:
            product_grade = good.find_element(By.CLASS_NAME, 'b-list-advert-base__item-attr')
            grade = product_grade.text.strip() if product_grade else 'NA'
        except:
            grade = 'NA'

        product = {
            'Name': name,
            'Price': price,
            'Description': description,
            'Image_Link': image,
            #'Location': location,
            'Grade': grade
        }
        jiji_data.append(product)
driver.quit() 
#always quit the driver
```
